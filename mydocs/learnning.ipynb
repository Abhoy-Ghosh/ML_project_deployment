# 📁 ROOT PROJECT FOLDER

```text id="0vk57c"
ML_projects/
```

👉 Main project directory

Contains:

* source code
* configs
* requirements
* setup files

---

# 📁 `src/`

```text id="y5lk4q"
src = source code
```

👉 All actual application logic lives here

This is industry standard.

Why?

```text id="f0pq9m"
Keeps business logic separate from random files
```

---

# 📁 `components/`

```text id="ccj90e"
Reusable ML building blocks
```

This is one of the MOST important folders.

---

## 🔹 `data_ingestion.py`

👉 Purpose:

```text id="spbrt5"
Collect / read data
```

Examples:

* CSV loading
* DB fetching
* API reading

---

### Why separate?

Because:

```text id="v74j2k"
Data source may change later
```

You don’t want ingestion logic mixed with training.

---

---

## 🔹 `data_transformation.py`

👉 Purpose:

```text id="xx2s7t"
Clean + preprocess data
```

Examples:

* missing value handling
* encoding
* scaling
* pipelines

---

### Why separate?

Because preprocessing is:

```text id="8av0aa"
independent of model training
```

---

---

## 🔹 `model_trainer.py`

👉 Purpose:

```text id="3qg7zq"
Train ML models
```

Examples:

* fit()
* evaluate()
* save model

---

### Why separate?

You may:

* change algorithms
* add GridSearchCV
* compare models

without touching preprocessing.

---

# 📁 `pipeline/`

This is the **orchestrator layer**

Very important concept.

---

## 🔹 `train_pipeline.py`

👉 Purpose:

```text id="6x3igj"
Runs complete training flow
```

Flow:

```text id="6nzwpv"
ingestion
→ transformation
→ training
→ save model
```

---

### Why needed?

Without pipeline:

```text id="yr22eu"
You manually run files ❌
```

With pipeline:

```text id="r68wbr"
One command runs everything ✅
```

---

---

## 🔹 `predict_pipeline.py`

👉 Purpose:

```text id="9f4lqz"
Take new user data → predict
```

Used during:

* deployment
* FastAPI
* Streamlit

---

### Real-world use

User enters:

```text id="c8kag4"
car details
```

Pipeline:

```text id="ph09p0"
transform → predict → output
```

---

# 📁 `logger.py`

One of the MOST important production files.

---

## Purpose

```text id="0g10xr"
Track application activity
```

Examples:

* model started training
* error occurred
* data loaded

---

## Why needed?

Without logging:

```text id="c0wnim"
Debugging becomes nightmare ❌
```

---

# 📁 `exception.py`

👉 Custom exception handling

Purpose:

```text id="d10wzc"
Generate readable professional errors
```

Instead of:

```text id="oc5mk7"
ValueError somewhere...
```

You get:

```text id="r3eqjk"
Error in model_trainer.py line 42
```

---

# 📁 `utils.py`

👉 Common helper functions

Examples:

* save_object()
* load_object()
* evaluate_model()

---

## Why separate?

Avoid duplicate code.

---

# 📁 `__init__.py`

Very important Python concept.

Purpose:

```text id="06nqec"
Makes folder behave like package
```

Allows:

```python id="r1dxna"
from src.components.data_ingestion import DataIngestion
```

---

# 📁 `requirements.txt`

👉 List of dependencies

Example:

```text id="dd6bzr"
pandas
numpy
scikit-learn
```

---

## Why needed?

So others can run:

```bash id="tt1b9w"
pip install -r requirements.txt
```

---

# 📁 `setup.py`

Advanced Python packaging file.

Purpose:

```text id="dbuggr"
Turn project into installable package
```

---

# 📁 `venv/`

👉 Virtual environment

Purpose:

```text id="9m51py"
Isolate project dependencies
```

---

# 📁 `.gitignore`

👉 Prevent unnecessary files from GitHub

Examples:

* venv
* logs
* cache

---

# 📁 `README.md`

👉 Project documentation

Explains:

* project
* setup
* usage

---

# 📁 `logs/`

👉 Stores log files

Generated by:

```python id="bhk4ic"
logging.basicConfig()
```
